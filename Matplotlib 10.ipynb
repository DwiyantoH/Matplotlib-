```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
#Sample DataSet
```

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
#Sample data ini memanfaatakan seed random yang dimiliki oleh numpy.  Pemanggilan fungsi np.random.normal menghasilkan bilangan acak yang akan membentuk distribusi normal
```

```python
plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

```python
#Boxplot dasarnuya terdiri dari suatu box/kotak.
```

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data,
           labels=['Data'],
           showfliers=True,
           flierprops={'markerfacecolor':'r', 'marker':'d'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Marker untuk outlier pada boxplot dalam matplotlib dikenal sebagai flier.
```

```python
#cara untuk horizontal Box Plot
```

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
np.random.seed(10)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(90, 20, 200)
data4 = np.random.normal(70, 25, 200)

data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horisontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
