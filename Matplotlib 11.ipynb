```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
#Pengaturan pada Violin Plot
```

```python
plt.violinplot(data,
              showextrema=False,
              showmeans=False,
              showmedians=False,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=False,
              showmedians=False,
              quantiles=None)

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
              showextrema=True,
              showmeans=False,
              showmedians=False,
              quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data, vert=False)

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data, vert=True)

plt.title('Horisontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
